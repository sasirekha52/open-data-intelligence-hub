#Question1. What is regression in machine learning, and when is it used?





Answer:
Regression is a machine learning technique used to predict continuous numerical values. We use it when the output is a number instead of a category. For example, predicting house prices, salary, temperature, or sales.

#Question2. What is the difference between simple linear regression and multiple linear regression?

Answer:
1. Simple Linear Regression uses only one independent variable to predict the output.

   Example: Predicting salary based only on years of experience.

2. Multiple Linear Regression uses two or more independent variables.

   Example:Predicting salary based on experience, education, skills, and age.

#Question3. What assumptions are made by linear regression?

Linear regression assumes:


*   There is a linear relationship between input and output.

*   The observations are independent.
*   The variance of errors remains constant.


*   Errors are normally distributed.

*   Independent variables should not be highly correlated with each other.






#Question4. What is the difference between a dependent variable and an independent variable?

The dependent variable is the value we want to predict.
The independent variables are the input features used to make that prediction.

Example:
If we predict house price:


*   House price → Dependent variable
*   Area, bedrooms, location → Independent variables



#Question5. What do the slope and intercept represent in a linear regression model?

Answer:
The slope tells us how much the output changes when the input increases by one unit.

The intercept is the predicted value when all input variables are zero.

Equation:
y = mx + c


*   m = slope
*   c = intercept



#Question6. What is the purpose of the cost function in regression?

Answer:
The cost function measures how far the predicted values are from the actual values.

The goal of training is to minimize this cost so that the predictions become more accurate.

#Question7. What is Mean Squared Error (MSE), and why is it commonly used?

Answer:
Mean Squared Error is the average of the squared differences between actual and predicted values.

It is widely used because it penalizes large errors more heavily, making the model focus on reducing big mistakes.

Lower MSE means better performance.

#Question8. What is the difference between MAE, MSE, and RMSE?

Answer:

MAE (Mean Absolute Error)


*   Average of absolute errors.

*   Easy to understand.
*  Less affected by outliers.

MSE (Mean Squared Error)

*  Squares the errors.

*   Gives more importance to large errors.
*   Sensitive to outliers.

RMSE (Root Mean Squared Error)


*   Square root of MSE.

*   Measured in the same unit as the target variable.
*   Easier to interpret than MSE.










#Question9. What is the R-squared score, and how should it be interpreted?

R-squared tells us how well the model explains the variation in the target variable.

Its value ranges from 0 to 1.


*   0 means the model explains nothing.
*   1 means perfect prediction.

For example, an R² score of 0.90 means the model explains 90% of the variation in the data.

#Question11. What is overfitting in a regression model?

Answer:
Overfitting happens when the model learns the training data too well, including its noise.

As a result, it performs very well on training data but poorly on new or unseen data.

#Question12. What is underfitting in a regression model?

Answer:
Underfitting happens when the model is too simple to capture the pattern in the data.

It performs poorly on both training and testing data.

#Question13. What is multicollinearity, and how does it affect regression?

Answer:
Multicollinearity occurs when two or more independent variables are highly correlated.

It makes it difficult for the model to determine the individual effect of each feature.

This can lead to unstable coefficients and unreliable predictions.

#Question14. How can multicollinearity be detected?

Answer:
Some common methods are:


*   Correlation matrix

*   Variance Inflation Factor (VIF)
*   Heatmap

A high VIF value usually indicates multicollinearity.



#Question15. What is regularization, and why is it used?

Answer:
Regularization is a technique used to reduce overfitting.

It adds a penalty to the model for having very large coefficients, helping the model generalize better on unseen data.

#Question16. What is the difference between Ridge, Lasso, and Elastic Net regression?

Answer:

Ridge Regression


*   Uses L2 regularization.
*   Reduces coefficient values but does not remove features.

Lasso Regression


*   Uses L1 regularization.
*  Can reduce some coefficients to zero, effectively removing less important features.

Elastic Net


*   Combines both Ridge and Lasso.
*   Works well when features are highly correlated.








#Question17. How does Lasso regression perform feature selection?

Answer:
Lasso adds an L1 penalty to the cost function.

As a result, it shrinks some coefficients completely to zero.

Features with zero coefficients are automatically removed, so Lasso performs feature selection.

#Question18. What is polynomial regression?

Answer:
Polynomial regression is used when the relationship between variables is not linear.

It adds polynomial terms like x², x³, etc., allowing the model to fit curved relationships.

#Question19. How do outliers affect a regression model?

Answer:
Outliers can significantly affect regression because they pull the regression line toward themselves.

This may reduce the model's accuracy.

We can handle outliers by detecting them using box plots or IQR and removing or treating them if necessary.

#Question20. How would you evaluate the performance of a regression model?

Answer:
I would evaluate a regression model using:


*   MAE – average prediction error.

*  MSE – gives higher penalty to large errors.
*   RMSE – error in the same unit as the target.


*  R-squared – shows how much variation the model explains.

*   Adjusted R-squared – useful when comparing models with different numbers of features.

I would also compare training and testing performance to check for overfitting or underfitting.
