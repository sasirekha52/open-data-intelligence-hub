```python
##Load Dataset
import pandas as pd

df = pd.read_csv("netflix_titles_nov_2019.csv")

print(df.head())
print(df.shape)
```

```python
##Data Cleaning
df.drop_duplicates(inplace=True)

df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

df.to_csv("cleaned_netflix.csv", index=False)
```

```python
##Basic Statistics
print(df.shape)
print(df.isnull().sum())
print(df['type'].value_counts())
```

```python
##Movies vs TV Shows
type_count = df['type'].value_counts()
print(type_count)
```

```python
##Top 10 Countries
country_count = df['country'].value_counts().head(10)

print(country_count)
```

```python
##Content Added by Year
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['year_added'] = df['date_added'].dt.year

year_trend = df['year_added'].value_counts().sort_index()
```

```python
##Visualizations
##Chart 1: Movies vs TV Show
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar')

plt.title("Movies vs TV Shows")
plt.savefig("content_type.png")
```

```python
##Chart 2: Top Countries
year_trend.plot(kind='line')

plt.title("Netflix Content Growth")
plt.savefig("release_years.png")
```

```python
##Chart 3: Content Added Over Years
year_trend.plot(kind='line')

plt.title("Netflix Content Growth")
plt.savefig("release_years.png")
```
