```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
# Load Dataset
df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")

df.head()
```

```python
# Total Sales
print("Total Sales:", df["Sales"].sum())
```

```python
# Total Profit
print("Total Profit:", df["Profit"].sum())
```

```python
# Sales by Region
sales_region = df.groupby("Region")["Sales"].sum()
sales_region
```

```python
plt.figure(figsize=(8,5))
sales_region.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
# Profit by Region
profit_region = df.groupby("Region")["Profit"].sum()

plt.figure(figsize=(8,5))
profit_region.plot(kind='bar', color='green')
plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```

```python
# Sales by Category
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
# Profit by Category
category_profit = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(8,5))
category_profit.plot(kind='bar')
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

```python
# Top 10 Products by Sales
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_products.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
# Top 10 Customers by Sales
top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_customers.plot(kind='bar')
plt.title("Top 10 Customers")
plt.xlabel("Customer")
plt.ylabel("Sales")
plt.show()
```

```python
# Ship Mode Count
plt.figure(figsize=(7,5))
sns.countplot(data=df, x="Ship Mode")
plt.title("Ship Mode Distribution")
plt.show()
```

```python
# Segment Count
plt.figure(figsize=(7,5))
sns.countplot(data=df, x="Segment")
plt.title("Customer Segment Distribution")
plt.show()
```

```python
# Correlation Heatmap
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Monthly Sales Trend
df["Order Date"] = pd.to_datetime(df["Order Date"])

monthly_sales = df.groupby(df["Order Date"].dt.month)["Sales"].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
