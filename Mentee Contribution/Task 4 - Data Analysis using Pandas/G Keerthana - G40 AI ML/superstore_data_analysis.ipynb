#Part F: Feature Engineering

```python
def sales_category(value):
    if value >= 1000:
        return 'High'
    elif value >= 500:
        return 'Medium'
    else:
        return 'Low'

df['sales_category'] = df['sales'].apply(sales_category)

df[['sales', 'sales_category']].head()
```

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['day_name'] = df['order_date'].dt.day_name()

df[['order_date', 'year', 'month', 'day_name']].head()
```

```python
df['profit_margin'] = df['profit'] / df['sales']

df[['sales', 'profit', 'profit_margin']].head()
```

```python
def discount_category(discount):
    if discount == 0:
        return 'No Discount'
    elif discount <= 0.2:
        return 'Low Discount'
    else:
        return 'High Discount'

df['discount_category'] = df['discount'].apply(discount_category)

df[['discount', 'discount_category']].head()
```

```python
def quantity_category(quantity):
    if quantity <= 2:
        return 'Low'
    elif quantity <= 5:
        return 'Medium'
    else:
        return 'High'

df['quantity_category'] = df['quantity'].apply(quantity_category)

df[['quantity', 'quantity_category']].head()
```

##Recommended Action

Investigate loss-making products and optimize pricing policies.

Higher discounts are associated with lower profits.

##Business Meaning

Large discounts can hurt the company's profitability.

##Recommended Action

Implement discount limits and evaluate promotional campaigns carefully.

Customer demand changes seasonally.
