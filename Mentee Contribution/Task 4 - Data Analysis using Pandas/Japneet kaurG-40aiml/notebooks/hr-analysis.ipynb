#Part F: Feature Engineering

Task F1: Employee Experience Category

```python
def experience_category(x):
    if x <= 5:
        return "Beginner"
    elif x <= 10:
        return "Intermediate"
    elif x <= 20:
        return "Experienced"
    else:
        return "Expert"

df["experience_category"] = df["totalworkingyears"].apply(experience_category)

df[["totalworkingyears","experience_category"]].head()
```

Task F2: Salary Category

```python
def income_category(x):
    if x < 5000:
        return "Low"
    elif x < 10000:
        return "Medium"
    else:
        return "High"

df["income_category"] = df["monthlyincome"].apply(income_category)

df[["monthlyincome","income_category"]].head()
```

Task F3: Promotion Status

```python
df["promotion_status"] = df["yearssincelastpromotion"].apply(
    lambda x: "Recently Promoted" if x <= 2 else "Not Recently Promoted"
)

df[["yearssincelastpromotion","promotion_status"]].head()
```

Overtime Category

```python
df["overtime_status"] = df["overtime"].apply(
    lambda x: "Extra Hours" if x=="Yes" else "Normal Hours"
)

df[["overtime","overtime_status"]].head()
```

#Part H: Correlation Analysis

Task H1: Calculate Correlation Matrix

```python
# Select numerical columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Correlation matrix
correlation_matrix = numeric_df.corr()

correlation_matrix
```

Task H2: Correlation Heatmap

```python
plt.figure(figsize=(16,10))

sns.heatmap(
    correlation_matrix,
    cmap="coolwarm",
    annot=False,
    linewidths=0.5
)

plt.title("Correlation Heatmap", fontsize=16)

plt.show()
```

Task H3: Find Highly Correlated Variables

```python
corr_pairs = correlation_matrix.unstack()

corr_pairs = corr_pairs.sort_values(ascending=False)

corr_pairs.head(20)
```
