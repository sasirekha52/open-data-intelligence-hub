# Feature Classification

# Numerical Data Preprocessing Pipeline

```python
numeric_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

Missing values in numerical features will be replaced using the median.
After imputation, numerical features will be standardized using StandardScaler.
Combining preprocessing steps into a pipeline ensures that the same transformations are consistently applied to both the training and testing datasets.
This approach also reduces the risk of data leakage and makes the preprocessing workflow more organized and reproducible.

# Categorical Data Preprocessing Pipeline

```python
categorical_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

Missing values in categorical features are handled using the most frequent category (mode).
Categorical variables are transformed into numerical format using One-Hot Encoding.
handle_unknown="ignore" improves model robustness by preventing errors when new or unseen categories appear in the test dataset.
This preprocessing ensures that machine learning models can interpret categorical data effectively.

Step 6.4: Column Transformer

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_pipeline, numeric_features),
        ("cat", categorical_pipeline, categorical_features)
    ]
)
```

Explanation

Ye automatically:

Numerical columns → Numerical Pipeline
Categorical columns → Categorical Pipeline

Step 7: Final Random Forest Pipeline

```python
model_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(
        n_estimators=100,
        random_state=42
    ))
])
```

Step 8: Train the Model

```python
model_pipeline.fit(X_train, y_train)
```

The pipeline successfully combines preprocessing and model training in a single workflow.
Both numerical and categorical features are automatically transformed before training.
The model used here is RandomForestClassifier, which is powerful for classification tasks like churn prediction.
Training the pipeline ensures that the same preprocessing steps will automatically be applied during prediction, preventing data leakage.
The structure improves code reusability, simplicity, and reliability.
