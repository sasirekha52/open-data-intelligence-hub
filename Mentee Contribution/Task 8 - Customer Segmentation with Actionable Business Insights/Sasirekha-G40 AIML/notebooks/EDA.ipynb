```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

pd.set_option("display.max_columns", None)
```

```python
engineered = [

    "Overall_Service_Score",

    "Comfort_Score",

    "Digital_Experience",

    "Customer_Value",

    "Travel_Experience",

    "Total_Delay"

]

train_df[engineered].hist(
    figsize=(15,10),
    bins=25
)

plt.tight_layout()

plt.show()
```

```python
service_columns = [

    "Inflight wifi service",

    "Food and drink",

    "Online boarding",

    "Seat comfort",

    "Inflight entertainment",

    "On-board service",

    "Leg room service",

    "Baggage handling",

    "Checkin service",

    "Inflight service",

    "Cleanliness"

]

plt.figure(figsize=(14,6))

train_df[service_columns].mean().sort_values().plot(kind="bar")

plt.title("Average Service Ratings")

plt.ylabel("Average Rating")

plt.show()
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(

    train_df.corr(),

    cmap="coolwarm",

    annot=False

)

plt.title("Correlation Heatmap")

plt.show()
```

```python
corr = train_df.corr()["satisfaction"]

corr = corr.sort_values()

plt.figure(figsize=(8,10))

corr.plot(kind="barh")

plt.title("Feature Correlation with Satisfaction")

plt.show()
```

```python
pair_features = [

    "Overall_Service_Score",

    "Comfort_Score",

    "Digital_Experience",

    "Flight Distance",

    "Total_Delay",

    "satisfaction"

]

sns.pairplot(

    train_df[pair_features],

    hue="satisfaction"

)

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(

    data=train_df,

    x="Age",

    y="Customer_Value",

    hue="satisfaction",

    alpha=0.6

)

plt.title("Age vs Customer Value")

plt.show()
```
