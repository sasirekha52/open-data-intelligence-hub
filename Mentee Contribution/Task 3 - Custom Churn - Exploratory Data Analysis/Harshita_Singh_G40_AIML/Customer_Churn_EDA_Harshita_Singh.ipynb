#Univariate Analysis

#Visualization 1: Churn Count Plot

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize=(6,4))

sns.countplot(x='Churn', data=df)

plt.title("Customer Churn Distribution")
plt.show()
```

#Observation 1-
The dataset is heavily imbalanced. Churn = 1.0 (churned customers) dominates , while Churn = 0.0 (retained) is very small (~10,000). This is a significant class imbalance that would affect any ML model built on this data.

#Visualisation2 : Churn percentage pie chart

```python
df['Churn'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title("Customer Churn Percentage")
plt.ylabel("")
plt.show()
```

#Observation2
The pie chart confirms that customer churn is extremely high. Nearly all customers in the dataset have churned, highlighting the need for effective retention strategies.

#Visualisation3 : Age Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Age'], bins=30)

plt.title("Age Distribution")
plt.show()
```

#Observation 3-
Age is fairly uniformly distributed between 18–65, with a slight bimodal pattern — slightly higher counts around ages 20–25 and 55–60, with minor dips in between. No strong age concentration exists, suggesting age alone may not be a key churn driver.

#Visualisation 4 : Tenure Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Tenure'], bins=30)

plt.title("Customer Tenure Distribution")
plt.show()
```

#Observation 4 -
Tenure (0–60 months) is also nearly uniform across all bins, meaning the dataset includes customers at all stages of their lifecycle equally. No particular tenure group dominates, suggesting tenure doesn't cluster churn risk on its own.

#Visualisation 5 :Total Spend Boxplot

```python
plt.figure(figsize=(8,5))

sns.boxplot(y=df['Total Spend'])

plt.title("Total Spend Boxplot")
plt.show()
```

#Observation 5-
 The median total spend is around 580-600 ,The IQR spans roughly around 450-750 indicating most customers spend in that range.The whiskers extend from ~100 to ~1000 with no visible outliers, suggesting a wide but evenly spread spending distribution.

#Bivariate Analysis with Churn

#Visualisation 6 : Gender VS Churn

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Gender',
    hue='Churn',
    data=df
)

plt.title("Gender vs Churn")
plt.show()
```

#Observation 6-
Both Female and Male groups show nearly identical churn patterns — the churned (1.0, orange) bar vastly overwhelms the retained (0.0, blue) bar for both genders. Gender does not appear to be a differentiating factor for churn.

#Visualisation 7 : Subscription Type VS Churn

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Subscription Type',
    hue='Churn',
    data=df
)

plt.title("Subscription Type vs Churn")
plt.show()
```

#Observation7-
All three subscription types — Standard, Basic, and Premium — show similarly high churn (1.0) relative to non-churn (0.0). No subscription tier appears to retain customers significantly better, suggesting subscription type has little influence on churn.

#Visualisation 8 : Contract Length VS Churn

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Contract Length',
    hue='Churn',
    data=df
)

plt.title("Contract Length vs Churn")
plt.show()
```

#Observation8 -
Annual, Monthly, and Quarterly contracts all show predominantly high churn. Interestingly, Monthly contracts appear to have a slightly higher churned count than Annual or Quarterly, but the difference is modest. Contract length doesn't strongly differentiate churn behavior here.

#Correlation Analysis

#Visualisation 9 : Correlation Heatmap

```python
plt.figure(figsize=(10,6))

numeric_cols = df.select_dtypes(include='number')

sns.heatmap(
    numeric_cols.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

#Observation 9 -
Most features show very weak correlations with Churn. The strongest correlations with Churn are Support Calls (0.22) — the highest positive correlation, suggesting customers who call support more tend to churn more — and Total Spend (−0.16), indicating higher spenders are slightly less likely to churn. All other variables (Age, Tenure, Usage Frequency, Payment Delay, Last Interaction) have near-zero correlation with Churn.

#Visualisation 10 : Total Spend VS Tenure

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='Tenure',
    y='Total Spend',
    hue='Churn',
    data=df
)

plt.title("Tenure vs Total Spend")
plt.show()
```

#Observation 10 -
Churned customers (orange, 1.0) cluster in the lower spend range, while retained customers (blue, 0.0) tend to occupy the higher spend range. This aligns with the correlation heatmap finding that higher total spend is associated with lower churn. Tenure shows no clear separating pattern — both groups span the full 0–60 month range evenly.

# Final Insights
1. The overall customer churn rate is approximately 95.89%, indicating a significant customer retention challenge.

2. Customers with Basic subscription plans experience the highest churn rate compared to Premium and Standard subscribers.

3. Monthly contract customers exhibit the highest churn rate, reaching nearly 100%, making contract length one of the strongest indicators of churn.

4. Support Calls show the strongest positive relationship with churn, suggesting that customers facing frequent issues are more likely to leave.

5. Payment Delay is positively associated with churn, indicating that customers experiencing payment-related difficulties have a higher likelihood of leaving.

6. Total Spend has a negative relationship with churn, suggesting that higher-value customers are generally more loyal and less likely to churn.

7. Customer Tenure also shows a negative relationship with churn, meaning long-term customers tend to stay with the company.

8. The analysis demonstrates that customer retention is influenced by contract structure, customer service interactions, spending behavior, and customer engagement.

# Business Recommendations
1. Promote Long-Term Contracts

   Encourage customers to switch from monthly plans to annual or quarterly contracts through discounts, loyalty rewards, and exclusive benefits.

2. Improve Customer Support Services

   Since support calls show the strongest positive relationship with churn, the company should improve issue resolution speed, customer service quality, and support efficiency.

3. Reduce Payment Delays

   Implement automated payment reminders, flexible payment methods, and proactive billing support to minimize payment-related churn.

4. Strengthen Customer Retention Programs

   Develop loyalty programs, referral incentives, and personalized offers for existing customers to improve customer satisfaction and retention.

5. Focus on High-Risk Customers

   Identify customers with frequent support calls, payment delays, low spending levels, or short contract lengths and target them with retention campaigns.

6. Increase Customer Engagement

   Encourage customers to use services more frequently through personalized communication, educational content, and promotional campaigns.

7. Reward High-Value Customers

   Provide exclusive benefits and recognition programs for customers with high spending and long tenure to maintain loyalty.

# Conclusion
This exploratory data analysis examined customer churn patterns using a dataset containing more than 260,000 customer records. The dataset was cleaned, analyzed, and visualized to identify factors influencing customer churn.

The analysis revealed an exceptionally high churn rate of approximately 95.89%. Contract length, support calls, payment delays, total spending, and customer tenure were identified as important factors influencing churn behavior. Monthly contract customers and customers requiring frequent support demonstrated the highest churn tendencies, while customers with higher spending and longer tenure showed stronger retention.

The findings suggest that improving customer support, encouraging long-term contracts, reducing payment-related issues, and implementing targeted retention strategies can help reduce customer churn and improve overall business performance.

Overall, the analysis provides valuable insights that can assist organizations in developing effective customer retention strategies and enhancing long-term customer relationships.
