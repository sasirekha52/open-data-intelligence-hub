# Key Findings

1. The dataset contains 7043 customer records.

2. The overall churn rate is approximately 26.5%.

3. Customers with month-to-month contracts show the highest churn rate.

4. Customers with lower tenure are more likely to leave the company.

5. Higher monthly charges are associated with increased churn.

6. Contract type is one of the strongest factors affecting customer retention.

# Business Recommendations

1. Encourage customers to switch from month-to-month contracts to long-term contracts.

2. Introduce loyalty programs for customers with low tenure.

3. Offer discounts and personalized retention offers to customers at high risk of churn.

4. Improve customer engagement during the first few months of service.

5. Monitor customers with high monthly charges and provide suitable service plans.

# Conclusion

The analysis shows that tenure, contract type, and monthly charges are important factors influencing customer churn. Businesses can reduce churn through targeted retention strategies and long-term customer engagement programs.
