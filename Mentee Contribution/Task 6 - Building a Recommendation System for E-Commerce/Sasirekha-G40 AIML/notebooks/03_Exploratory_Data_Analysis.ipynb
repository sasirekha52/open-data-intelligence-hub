```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

plt.style.use("ggplot")
```

```python
df = pd.read_csv("../data/processed/ecommerce_processed.csv")
```

```python
df.head()
```

```python
print("Rows and Columns:")
print(df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"], bins=15)

plt.title("Customer Age Distribution")

plt.xlabel("Age")

plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="Gender", data=df)

plt.title("Gender Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x="Category", data=df)

plt.xticks(rotation=45)

plt.title("Product Categories")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="Purchase_Status", data=df)

plt.title("Purchase Status")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(x="Rating", data=df)

plt.title("Ratings")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="Discount_Applied", data=df)

plt.title("Discount Applied")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Browsing_Time"], bins=30)

plt.title("Browsing Time")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Total_Spending"], bins=30)

plt.title("Total Spending")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Price"], bins=20)

plt.title("Price Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Age",
    y="Total_Spending",
    data=df
)

plt.title("Age vs Spending")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Price",
    y="Rating",
    data=df
)

plt.title("Price vs Rating")

plt.show()
```

```python
plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=["number"])

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Price"])

plt.title("Price Outliers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Browsing_Time"])

plt.title("Browsing Time Outliers")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    x="Category",
    y="Total_Spending",
    data=df
)

plt.xticks(rotation=45)

plt.title("Average Spending by Category")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.barplot(
    x="Gender",
    y="Purchase_Status",
    data=df
)

plt.title("Purchase Rate by Gender")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    x="Category",
    y="Purchase_Status",
    data=df
)

plt.xticks(rotation=45)

plt.title("Purchase Rate by Category")

plt.show()
```
