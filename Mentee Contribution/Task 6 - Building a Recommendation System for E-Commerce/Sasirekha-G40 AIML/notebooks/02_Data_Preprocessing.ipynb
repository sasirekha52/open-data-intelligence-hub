```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"C:\Users\ELCOT\Desktop\Ecommerce-Recommendation-System\data\raw\ecommerce_data.csv.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df["User_ID"] = range(100001,100001+len(df))
```

```python
df[["User_ID"]].head()
```

```python
df["Product_ID"] = np.random.randint(5001,5301,size=len(df))
```

```python
df[["Product_ID"]].head()
```

```python
category_map = {
1:"Electronics",
2:"Fashion",
3:"Home",
4:"Books",
5:"Sports",
6:"Beauty",
7:"Toys",
8:"Grocery"
}

df["Category"] = df["TrafficType"].map(category_map)
```

```python
df["Category"] = df["Category"].fillna("Others")
```

```python
df["Category"].value_counts()
```

```python
df["Browsing_Time"] = (
df["Administrative_Duration"] +
df["Informational_Duration"] +
df["ProductRelated_Duration"]
)
```

```python
df["Browsing_Time"].head()
```

```python
df["Previous_Purchases"] = (
df["ProductRelated"]/5
).astype(int)
```

```python
np.random.seed(42)

df["Price"] = np.random.randint(500,5000,len(df))
```

```python
df["Price"].head()
```

```python
df["Rating"] = np.where(
df["PageValues"]>20,
5,
np.where(
df["PageValues"]>10,
4,
np.where(
df["PageValues"]>5,
3,
2
)
)
)
```

```python
df["Rating"].value_counts()
```

```python
df["Cart_Addition"] = np.where(
df["PageValues"]>0,
1,
0
)
```

```python
df["Purchase_Status"] = df["Revenue"].astype(int)
```

```python
df["Purchase_Status"].value_counts()
```

```python
np.random.seed(42)

df["Age"] = np.random.randint(
18,
60,
len(df)
)
```

```python
np.random.seed(42)

df["Gender"] = np.random.choice(
["Male","Female"],
len(df)
)
```

```python
location_map = {
1:"North",
2:"South",
3:"East",
4:"West",
5:"Central",
6:"North-East",
7:"International",
8:"Urban",
9:"Rural"
}

df["Location"] = df["Region"].map(location_map)

df["Location"] = df["Location"].fillna("Unknown")
```

```python
df["Discount_Applied"] = np.where(
df["SpecialDay"]>0,
1,
0
)
```

```python
df["Total_Spending"] = (
df["Price"]*
df["Purchase_Status"]
)
```

```python
final_df = df[[
"User_ID",
"Product_ID",
"Category",
"Price",
"Rating",
"Browsing_Time",
"Previous_Purchases",
"Cart_Addition",
"Purchase_Status",
"Age",
"Gender",
"Location",
"Discount_Applied",
"Total_Spending"
]]
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df.to_csv(
"../data/processed/ecommerce_processed.csv",
index=False
)
```
